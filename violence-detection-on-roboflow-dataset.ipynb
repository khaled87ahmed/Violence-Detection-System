```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the results.csv file
csv_path = "/kaggle/working/runs/violence_yolov82/results.csv"
df = pd.read_csv(csv_path)

# Optional: clean column names (remove leading/trailing spaces)
df.columns = df.columns.str.strip()

# Plotting
plt.figure(figsize=(15, 6))

# Precision and Recall
plt.subplot(1, 3, 1)
plt.plot(df['metrics/precision(B)'], label='Precision')
plt.plot(df['metrics/recall(B)'], label='Recall')
plt.title('Precision & Recall')
plt.xlabel('Epoch')
plt.ylabel('Score')
plt.legend()
plt.grid()

# mAP
plt.subplot(1, 3, 2)
plt.plot(df['metrics/mAP50(B)'], label='mAP@0.5')
plt.plot(df['metrics/mAP50-95(B)'], label='mAP@0.5:0.95')
plt.title('mAP Metrics')
plt.xlabel('Epoch')
plt.ylabel('Score')
plt.legend()
plt.grid()

# Loss
plt.subplot(1, 3, 3)
plt.plot(df['train/box_loss'], label='Train Box Loss')
plt.plot(df['val/box_loss'], label='Val Box Loss')
plt.title('Box Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
